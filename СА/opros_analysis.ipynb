```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv').drop('Отметка времени', 1)
df.head()
```

```python
col_names = [
    'Скажи мне, ты парень или девушка?',
    'Комфортно ли тебе пребывать в состоянии одиночества?',
    'Как ты отнесешься к тому, что с моей помощью ты сможешь заказать доставку чего-нибудь вкусненького ?',
    'Пользовался ли ты раньше голосовым помощником? (Алиса, Siri, Cortana и др.)'
]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,15))
for i, col_name in enumerate(col_names):
    df[col_name].value_counts().plot.pie(autopct="%.0f%%", fontsize=12, ax=axes[i // 2, i % 2])
```
